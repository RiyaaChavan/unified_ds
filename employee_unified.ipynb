```python
median_year = df['NumCompaniesWorked'].median()
df['NumCompaniesWorked'] = df['NumCompaniesWorked'].fillna(median_year)
print("\nDataFrame after filling NaN with median:")

```

```python
median_year = df['JobSatisfaction'].median()
df['JobSatisfaction'] = df['JobSatisfaction'].fillna(median_year)
print("\nDataFrame after filling NaN with median:")

```

```python
median_year = df['EnvironmentSatisfaction'].median()
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].fillna(median_year)
print("\nDataFrame after filling NaN with median:")
```

```python
median_year = df['WorkLifeBalance'].median()
df['WorkLifeBalance'] = df['WorkLifeBalance'].fillna(median_year)
print("\nDataFrame after filling NaN with median:")
```

```python
median_year = df['TotalWorkingYears'].median()
df['TotalWorkingYears'] = df['TotalWorkingYears'].fillna(median_year)
print("\nDataFrame after filling NaN with median:")
```
